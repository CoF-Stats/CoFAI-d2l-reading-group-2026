> Why is the $K$-fold cross-validation error estimate biased?

Since the splits are even, we can rewrite the $e_{CV}$ estimate as instead

$$
\begin{aligned}
e_{CV} &= \frac{1}{K}\sum_{k = 1}^{K} \sum_{i = 1}^{N/K} (y_{ik} - \hat{y}_{ik} | \mathbf{X}_{-k}, \mathbf{y}_{-k})^2\\
&= \sum_{i = 1}^N (y_i^{(k)} - \hat{y}_i | \mathbf{X}_{-k}, \mathbf{y}_{-k})^2
\end{aligned}
$$

where I have "stacked" the error estimates into a vector of length $N$ and used the vertical bar $|$ notation to reflect that each fold's outputs are predicted using the remaining folds' inputs.
