### Exercise 3.6.5.5

> Why is the $K$-fold cross-validation error estimate biased?

Whether the CV error estimate is biased depends on the *target of inference* or *estimand*. Bates et al. [(2023)](https://pmc.ncbi.nlm.nih.gov/articles/PMC11412612/) show that CV error estimates the error of predictions over many different training sets of the same size. However, the more useful measure in practice is the prediction error of a model *trained on a given training set*. More precisely, let $E_{CV}$ be the cross validation error estimator,

$$
E_{CV} = \sum_{k = 1}^K \sum_{i = 1}^{N/K} L(y_{ik} - \hat{f}(\mathbf{x}_{ik}\ |\ \mathbf{X}_{-k}, \mathbf{y}_{-k}))
$$

where $L()$ is the loss function, $\mathbf{X}$ is the matrix of features, $\mathbf{y}$ is the response/output, and $\mathbf{X}_{-k}$ denotes the rows of $\mathbf{X}$ minus those in the $k^\text{th}$ fold. Because the training and validation sets are independent across each fold and selected at random, both the errors and the training sets can be viewed as random variables and the expected value of $E_{CV}$ can be shown to be

$$
\begin{aligned}
    \mathbb{E}[E_{CV}] &= \int\int \sum_{k = 1}^K \sum_{i = 1}^{N/K} L(y_{ik} - \hat{f}(\mathbf{x}_{ik}\ |\ \mathbf{X}_{-k}, \mathbf{y}_{-k})) P(X, Y) d\mathbf{X}d\mathbf{y}\\
    &= \text{tr}(\text{Cov}(\mathbf{\epsilon}))
\end{aligned}
$$

where $P(X, Y)$ is the joint distribution of features and response, $\text{tr}()$ is the matrix trace and $\mathbf{\epsilon}$ is the prediction error from a randomly selected target $y$ and random draw of features $\mathbf{x}_i$, trained on a random training set of size $N - N/K$. 

Because this quantity averages over all possible training sets $\mathbf{X}$, it *underestimates* the more practically useful error of interest, which is instead

$$
\mathbb{E}_{\tilde Y, \tilde{\mathbf{x}}|\mathbf{X},\mathbf{Y}}\left[ L(\tilde Y, \hat{f}(\tilde{\mathbf{x}} | \mathbf{X}, \mathbf{y})) \right]
$$

where the training set is no longer treated as a random variable. Thus, when we condition on a given training set, the finite sample error does not average out in the calculation, so the CV error estimate is biased downward, at least in reference to the actual prediction error of interest.

#### References

Bates S, Hastie T, Tibshirani R. Cross-validation: what does it estimate and how well does it do it? *J Am Stat Assoc*. 2024;119(546):1434-1445. doi: 10.1080/01621459.2023.2197686.
