> Design a binary code for a three-class classification problem, where each class occurs with equal probability. What is the "problem" with this code?

I'm not totally sure what they mean by "problem" here, but my guess is that distributions with equally-likely states have high entropy. So, by information theory, the best we could do is

$$
-\log_2\left( \frac{1}{3} \right) \approx 1.58 \le 2 \text{ bits}
$$

so we would need 2 bits to make this encoding. Something like $\{00, 01, 10\}$ for the three classes would work.

> Can you design a better code? *Hint*: what happens if we try to encode two independent observations? What if we encode $n$ observations jointly?

If we encode the unique possibilities of two independent observations, which include the summed outcomes

$$
\begin{cases}
\{1, 1, 0\} & \text{wp } 2/9 \\
\{1, 0, 1\} & \text{wp } 2/9 \\
\{0, 1, 1\} & \text{wp } 2/9 \\
\{2, 0, 0\} & \text{wp } 1/9 \\
\{0, 2, 0\} & \text{wp } 1/9 \\
\{0, 0, 2\} & \text{wp } 1/9 \\
\end{cases}
$$

then we can assign longer bit strings to the less likely outcomes and shorter bit strings to the more likely outcomes. This allows for, on average, a more efficient encoding. For example, we could use a prefix encoding, like

$$
\begin{cases}
\{1, 1, 0\} & \to 0 \\
\{1, 0, 1\} & \to 10 \\
\{0, 1, 1\} & \to 110 \\
\{2, 0, 0\} & \to 1110 \\
\{0, 2, 0\} & \to 11110 \\
\{0, 0, 2\} & \to 11111 \\
\end{cases}
$$

then, the average number of bits needed to jointly encode the two independent observations is $26/9 \approx 2.89$ which is less than the 3-bit naive approach of giving each outcome a unique string of 3 bits. This takes advantage of the fact that some outcomes are more likely than others.

```python
import numpy as np
import math
def entropy_multinom(n, p):
  k = len(p)
  h = 0.0

  # next term
  h -= n * np.sum(p * np.log2(p))

  # next term computes probabilities
  for i in range(k):
    for j in range(n):
      h += math.comb(n, j + 1) * p[i]**(j + 1) * (1 - p[i])**(j + 1) * np.log2(np.prod(np.arange(1, j + 1)))

  return h
```
